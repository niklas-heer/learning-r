## Vectorized operations

R can perform a number of operations out of the box on vectors!

* `+`
* `-`
* `*`
* `/`
* `>`
* `<`
* ...

```r
x <- rnorm(5)
x
```

Print it out...

```r
for (i in x) {
    print(i)
}
```

```r
print(x[1])
print(x[2])
print(x[3])
print(x[4])
print(x[5])
```

```r
# conventional for loop
for (j in 1:5) {
    print(x[j])
}
```

----------------

```r
N <- 100
a <- rnorm(N)
b <- rnorm(N)
```

### Vectorized approach

```r
c <- a * b
c
```

### De-vectorized approach

```r
d <- rep(NA, N)
for (i in 1:N) {
    d[i] <- a[i] * b[i]
}
d
```

Learning: In R the vectorized approach is much faster!

# Speed comparison

We have results about the performance of different languages and we want to display them in a nice way.

The data is from: https://github.com/niklas-heer/speed-comparison

```r
library(ggplot2)

df <- read.csv("data/data_v1.4.csv", header=TRUE, sep=",")
```

```r
ggplot(data = df,
       aes(x = name,
           y = median)) +
    geom_bar(stat = "identity")
```

Flip it...

```r
ggplot(data = df,
       aes(x = name,
           y = median)) +
    geom_bar(stat = "identity") +
    coord_flip()
```

Name the axes...

```r
ggplot(data = df,
       aes(x = name,
           y = median)) +
    geom_bar(stat = "identity") +
    coord_flip() +
    labs(x = "Languages", y = "Median time (ms)")
```

Sort it...

```r
?reorder()
```

```r
ggplot(data = df,
       aes(x = reorder(name, median),
           y = median)) +
    geom_bar(stat = "identity") +
    coord_flip() +
    labs(x = "Languages", y = "Median time (ms)")
```

Form fastest to slowest...

```r
ggplot(data = df,
       aes(x = reorder(name, -median),
           y = median)) +
    geom_bar(stat = "identity") +
    coord_flip() +
    labs(x = "Languages", y = "Median time (ms)")
```

Also show the accuracy...

```r
ggplot(data = df,
       aes(x = reorder(name, -median),
           y = median,
           fill = accuracy)) +
    geom_bar(stat = "identity") +
    coord_flip() +
    labs(x = "Languages", y = "Median time (ms)")
```

Add labels to the bars...

```r
ggplot(data = df,
       aes(x = reorder(name, -median),
           y = median,
           fill = accuracy)) +
    geom_bar(stat = "identity") +
    coord_flip() +
    labs(x = "Languages", y = "Median time (ms)") +
    geom_text(aes(label = format(df$median, digits=2)), hjust = -0.13)
```

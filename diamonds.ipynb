# Getting insights with R

The data we get from the [Super Data Science](https://www.superdatascience.com/rcourse/) website.

![diamonds](https://raw.github.com/niklas-heer/learning-r/master/.github/img/Diamond_Clarity_Chart.jpg "Diamond Chart")

What we are trying to prove is that diamonds should be priced higher if they are more clear.

```r
mydata <- read.csv("data/Mispriced-Diamonds.csv")
```

Let's activate the `ggplot2` package which we'll be using.

```r
library(ggplot2)
```

Now let's see what we can find out...

```r
ggplot(data=mydata, aes(x=carat, y=price)) + geom_point()
```

Let's add the clarity as the color.

```r
ggplot(data=mydata, aes(x=carat, y=price, colour=clarity)) + geom_point()
```

Let's make it a little more clear.

```r
ggplot(data=mydata, 
       aes(x=carat, y=price, colour=clarity)) + 
    geom_point(alpha=0.1)
```

Let's disregard the data points which are statistically not relevant and filter them out.

```r
ggplot(data=mydata[mydata$carat<2.5,], 
       aes(x=carat, y=price, colour=clarity)) + 
    geom_point(alpha=0.1)
```

Let's add some lines to see it more clearly.

```r
ggplot(data=mydata[mydata$carat<2.5,], 
       aes(x=carat, y=price, colour=clarity)) + 
    geom_point(alpha=0.5) +
    geom_smooth()
```
